```python
# prompt: /content/sample_data/california_housing_test.csv 파일을 읽어서 적당한 차트를 출력해줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 파일 읽기
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# 히스토그램으로 housing_median_age 분포 확인
plt.figure(figsize=(8, 6))
sns.histplot(df['housing_median_age'], kde=True)
plt.title('Distribution of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.show()

# 산점도로 median_income과 median_house_value 관계 확인
plt.figure(figsize=(8, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()

# 상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
